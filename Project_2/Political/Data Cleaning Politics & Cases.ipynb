```python
import pandas as pd
import csv
```

```python
polcal = pd.read_csv('/Users/EvelynJulian/Desktop/Metis/Project 2/Political/PoliticCal.csv')

def stringmod(x):
    stringx = str(x)
    if stringx[0] == '+':
        stringx = stringx[1:-1]
    else:
        stringx = stringx[0:-1]
    stringx = float(stringx)
    return(stringx)

polcal = polcal.drop([" '1'", " '2'", " '3'", " '4'"," '6'", ' "7"'], axis=1)
polcal = polcal.rename(columns={" '5'": 'Dem-Rep %'})
polcal['Dem-Rep %'] = polcal['Dem-Rep %'].apply(stringmod)
polcal['State'] = "California"

cascal = pd.read_csv('/Users/EvelynJulian/Desktop/Metis/Project 2/Cases/CaseCal.csv')
mergecal = pd.concat([polcal, cascal], axis=1)
mergecal = mergecal.drop(['State1', 'County1', 'Deaths'], axis = 1)
```

```python
polpenn = pd.read_csv('/Users/EvelynJulian/Desktop/Metis/Project 2/Political/PoliticPenn.csv')

polpenn['Dem-Rep %'] = round((100 * (polpenn['Democratic']-polpenn['Republican'])/polpenn['Total']), 1)

polpenn = polpenn.drop(['Democratic', 'Republican', 'Total'], axis =1)
polpenn['State'] = "Pennsylvania"

caspenn = pd.read_csv('/Users/EvelynJulian/Desktop/Metis/Project 2/Cases/CasePenn.csv')
mergepenn = pd.concat([polpenn, caspenn], axis=1)
mergepenn = mergepenn.drop(['State1', 'County1', 'Deaths'], axis = 1)
```
