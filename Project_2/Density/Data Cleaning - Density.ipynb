```python
denvir.loc[-1] = [133.0, '806.0/sq mi', 'Emporia, VA / 420']
denvir.index = denvir.index + 1
denvir = denvir.sort_index()
df = [dencal, denpenn, dentex, dengeo, denvir, denken, denkan, denill]

x = [den.shape for den in df]

denvir
```

```python
df = pd.concat(df).reset_index()
df = df.drop(columns=['index', 'Rank'], axis = 1)

def popshave(x):
    pos = x.index(',')
    return x[:pos]

df['County'] = df['County / Population'].apply(popshave)
df = df.rename(columns = {'Population Density ▼': 'Population Density'})

def tostate(x):
    if 'CA' in x:
        return 'California'
    elif 'GA' in x:
        return 'Georgia'
    elif 'IL' in x:
        return 'Illinois'
    elif 'KS' in x:
        return 'Kansas'
    elif 'KY' in x:
        return 'Kentucky'
    elif 'TX' in x:
        return 'Texas'
    elif 'PA' in x:
        return 'Pennsylvania'
    elif 'VA' in x:
        return 'Virginia'
    else:
        return 0

def topop(x):
    x = x[:-6]
    x = x.replace(',', '')
    x = float(x)
    return x 

    
df['State'] = df['County / Population'].apply(tostate)
df['Population Density'] = df['Population Density'].apply(topop)
df = df.drop(columns=['County / Population'], axis = 1)
dfmega = df.sort_values(by=['State','County'], ascending= True)
dfmega
```

```python
dfmega.to_csv(r'MegaDensity.csv')
```
