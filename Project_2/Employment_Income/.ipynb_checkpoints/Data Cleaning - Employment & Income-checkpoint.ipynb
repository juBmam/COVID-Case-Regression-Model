```python
import pandas as pd
import csv
```

```python
uncal = pd.read_csv('/Users/juBmam/Desktop/Metis/Project_2/Employment_Income/UnemploymentReport_CA.csv')
unpenn = pd.read_csv('/Users/juBmam/Desktop/Metis/Project_2/Employment_Income/UnemploymentReport_PA.csv')
ungeo = pd.read_csv('/Users/juBmam/Desktop/Metis/Project_2/Employment_Income/UnemploymentReport_GA.csv')
unill = pd.read_csv('/Users/juBmam/Desktop/Metis/Project_2/Employment_Income/UnemploymentReport_IL.csv')
unkan = pd.read_csv('/Users/juBmam/Desktop/Metis/Project_2/Employment_Income/UnemploymentReport_KA.csv')
unken = pd.read_csv('/Users/juBmam/Desktop/Metis/Project_2/Employment_Income/UnemploymentReport_KY.csv')
untex = pd.read_csv('/Users/juBmam/Desktop/Metis/Project_2/Employment_Income/UnemploymentReport_TX.csv')
unvir = pd.read_csv('/Users/juBmam/Desktop/Metis/Project_2/Employment_Income/UnemploymentReport_VA.csv')

df = [uncal, unpenn, ungeo, unill, unkan, unken, untex, unvir]

drop_col = [un.drop(['Unnamed: 0', 'Unemployment Rate (percent)', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 12'], axis=1) for un in df]
rena_col = [un.rename(columns={'Unnamed: 1': 'County', 'Unnamed: 10': 'Unemployment Rate (2019)', 'Unnamed: 11': "Median Household Income (2019)"}) for un in drop_col]
```

```python
rena_col[0] = rena_col[0].drop([0, 1, 60, 61])
rena_col[1] = rena_col[1].drop([0, 1, 69, 70])
rena_col[2] = rena_col[2].drop([0, 1, 161, 162, 163])
rena_col[3] = rena_col[3].drop([0, 1, 104, 105])
rena_col[4] = rena_col[4].drop([0, 1, 107, 108])
rena_col[5] = rena_col[5].drop([0, 1, 122, 123])
rena_col[6] = rena_col[6].drop([0, 1, 256, 257, 258])
rena_col[7] = rena_col[7].drop([0, 1, 135, 136, 137])
```

```python
dfmega = pd.concat(rena_col).reset_index()

def tostate(x):
    x = str(x)
    if 'CA' in x:
        return 'California'
    elif 'GA' in x:
        return 'Georgia'
    elif 'IL' in x:
        return 'Illinois'
    elif 'KS' in x:
        return 'Kansas'
    elif 'KY' in x:
        return 'Kentucky'
    elif 'TX' in x:
        return 'Texas'
    elif 'PA' in x:
        return 'Pennsylvania'
    elif 'VA' in x:
        return 'Virginia'
    else:
        return 0
    
def countyclean(x):
    x = str(x)[:-4]
    if 'County' in x:
        return x[:-7]
    return x

def incomeclean(x):
    number = str(x)
    number = (number[1:])
    number = number.replace(',', '') 
    indexdot = number.find('.')
    if indexdot != -1:
        number = number[:indexdot]
    return (number)       

dfmega['State'] = dfmega['County'].apply(tostate)
dfmega['County'] = dfmega['County'].apply(countyclean)
#dfmega['Median Household Income (2019)'] = dfmega['Median Household Income (2019)'].apply(incomeclean)
dfmega = dfmega.sort_values(by=['State','County'], ascending= True)
dfmega

```

```python
dfmega.to_csv(r'MegaEmpInc.csv')
```
